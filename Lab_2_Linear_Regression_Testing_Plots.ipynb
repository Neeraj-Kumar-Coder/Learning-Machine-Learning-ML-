```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("./deeplearning.mplstyle")
```

```python
# x_train is the input variable (size in 1000 sqaure feet)
# y_train is the target (price in 1000 dollars)

x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
# m is the number of training examples
# .shape will return the number of elements in the corresponding dimension as a tupe
print(f"x_train.shape: {x_train.shape}")
m = x_train.shape[0]
print(f"Number of training examples are: {m}")
```

```python
print("Training dataset is as follows")
for index in range(0, m):
    print(f"(x^({index}), y^({index})) = ({x_train[index]}, {y_train[index]})")
```

```python
# Plotting the data points
plt.scatter(x_train, y_train, marker='x', c='r')

plt.title("House prices")
plt.xlabel("Size of house (in 1000 square feet)")
plt.ylabel("Price of house (in 1000 dollar)")
plt.show()
```

```python
# Plotting the linear regression function (y = wx + b)
w = 200
b = 100
print(f"w = {w}")
print(f"b = {b}")
```

```python
def compute_model_output(x, w, b):
    """
    Args:
        x (ndarray(m,)) : Data, m samples
        w : weigth
        b : bias

    Returns:
        y (ndarray(m,)) : target values
    """

    m = np.shape(x)[0]
    y_wb = np.zeros(m)
    for index in range(0, m):
        y_wb[index] = w * x[index] + b

    return y_wb
```

```python
tmp_f_wb = compute_model_output(x_train, w, b)

# Plotting the predictions
plt.plot(x_train, tmp_f_wb, c='b', label="Our prediction")

# Plotting the original data points
plt.scatter(x_train, y_train, marker='x', c='r', label="Actual value")

# Setting title and labels
plt.title("House Prices")
plt.xlabel("Size (in 1000 sqaure feet)")
plt.ylabel("Price (in 1000 dollar)")
plt.legend()
plt.show()
```
